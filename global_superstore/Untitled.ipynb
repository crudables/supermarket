```python
import pandas as pd
```

```python
store = pd.read_csv('Global-Superstore.csv', encoding='unicode_escape') # encoding= 'unicode_escape' was used to solve the problem of inability of utf-8 to decode character
```

```python
store.head()
```

```python
store.columns
```

```python
store.info()
```

```python
store.drop(columns=["Row ID","Order ID","Customer ID","Customer Name","Postal Code"], inplace=True)
```

```python
store.columns = store.columns.map(lambda x:x.replace(" ","_"))
store.columns = store.columns.map(lambda x:x.lower())
```

```python
store["Order_Date"] = pd.to_datetime(store["Order_Date"])
store["Ship_Date"] = pd.to_datetime(store["Ship_Date"])
```

```python
store.sample(10)
```

```python
# The most expensive item sold over the period
store[store.sales ==store.sales.max()][["product_name","market","sales","profit","discount"]]
```

```python
# sale of the most expensive item was at a loss.
```

```python
store[store.sales ==store.sales.min()][["product_name","market","sales","profit","discount"]]
```
